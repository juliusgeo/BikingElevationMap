```python
import geopandas as gpd
import requests
import time
from shapely.geometry import Polygon
import geopandas_osm.osm
import numpy as np
import descartes
%matplotlib inline
```

```python
leftbound = -75.5346
rightbound = -75.4986
upperbound = 40.1448
lowerbound = 40.1215
bbox = [(leftbound,lowerbound ), (rightbound,lowerbound), (rightbound,upperbound), (leftbound,upperbound)]
df = geopandas_osm.osm.query_osm('way', Polygon(bbox), recurse='down')
df.plot()
```

```python
arr=["residential", "primary", "secondary","tertiary","unclassified", "cycleway"]
df=df[df.highway.isin(arr)]
df.plot()
```

```python
newdata = gpd.GeoDataFrame()
newdata["street name"]=None
newdata["geometry"]=None
```

```python
all_coords_list = []
for geometry in df["geometry"]:
    for coord in geometry.coords:
        all_coords_list.append(coord)
```

```python
def is_intersection(val):
    num_matches = 0
    for i in all_coords_list:
        if val == i:
            num_matches +=1
    
    if(num_matches > 1):
        return True
    else:
        return False

def get_elev_diff(point1, point2):
        rqstring = "https://maps.googleapis.com/maps/api/elevation/json?locations="+str(point1[0])+","+str(point1[0])
        elevation=requests.get(rqstring)
        elev1 = elevation.json()["results"][0]["elevation"]
        time.sleep(.1)
        rqstring = "https://maps.googleapis.com/maps/api/elevation/json?locations="+str(point2[0])+","+str(point2[0])
        elevation=requests.get(rqstring)
        elev2 = elevation.json()["results"][0]["elevation"]
        return elev1-elev2
```

```python
elev_diff_list = np.load("philly_elev_diffs.npy")
elev_diff_list = [i*10 for i in elev_diff_list]
```

```python
new_index = -1
elev_diff_index = 0
for index, name, geometry in zip(df.index.values, df["name"],df["geometry"]):
    new_index +=1
    #print "Street name: ",name 
    coords = geometry.coords
    
    coords_list = [i for i in coords]
    
    diff_list = [coords[0]]
    index=0
    for coord in coords[1:-1]:
        index+=1
        if(is_intersection(coord)):
            #print "true"
            diff_list.append(index)
        
    diff_list.append(coords[-1])
    
    prev_index = 0
    #print(diff_list)
    seg_arr =[]
    if(len(diff_list) > 2):
        for i in diff_list[1:-1]:
            print "section from point ", coords[prev_index], "to point", coords[i], ":",
            elev_diff=round(get_elev_diff(coords[prev_index], coords[i])*.00001,7)
            elev_diff_list.append(elev_diff)
            
            #elev_diff=elev_diff_list[elev_diff_index]
            #elev_diff_index+=1
            seg_arr = coords_list[prev_index:i]
            if(elev_diff < 0):
                seg_arr.append((seg_arr[0][0],seg_arr[0][1]-elev_diff))
            else:
                seg_arr.append((seg_arr[-1][0],seg_arr[-1][1]-elev_diff))
            prev_index = i
            #print(elev_diff, seg_arr)
    else:
        print "section from point ", diff_list[0], "to point", diff_list[1], ":",
        elev_diff=round(get_elev_diff(diff_list[0], diff_list[1])*.00001,7)
        elev_diff_list.append(elev_diff)
        
        #elev_diff=elev_diff_list[elev_diff_index]
        #elev_diff_index+=1
        
        seg_arr = coords_list  
        if(elev_diff < 0):
                seg_arr.append((seg_arr[0][0],seg_arr[0][1]-elev_diff))
        else:
                seg_arr.append((seg_arr[-1][0],seg_arr[-1][1]-elev_diff))
        #print(elev_diff, seg_arr)
    #print "New seg_arr", seg_arr
    
    if(len(seg_arr) > 2 and elev_diff > .00001):
        poly = Polygon(seg_arr)
        newdata.loc[new_index, "geometry"] = poly
        newdata.loc[new_index, "street name"] = name
```
